```python

"""
Does phi(theta) = a0 + R*cos(3*theta) give Z3 structure
and Koide relation? Full numerical + analytic check.
"""
import numpy as np

TH = np.array([0., 2*np.pi/3, 4*np.pi/3])  # Z3 points

print("=" * 60)
print("1. cos(3θ) EVALUATED AT Z3 POINTS")
print("=" * 60)
print("""
  cos(3 × 0)      = cos(0)    = 1
  cos(3 × 2π/3)   = cos(2π)   = 1
  cos(3 × 4π/3)   = cos(4π)   = 1

  → ALL THREE give the same value!
  → cos(3θ) at Z3 points is a CONSTANT.
  → No mass difference between generations.
""")

# Numeric verification
for i, th in enumerate(TH):
    print(f"  cos(3 × {th:.4f}) = cos({3*th:.4f}) = {np.cos(3*th):.6f}")

print()
print("=" * 60)
print("2. KOIDE RATIO WITH cos(3θ) ONLY")
print("=" * 60)

# phi = a0 + R*cos(3*theta)
a0, R = 10.0, np.sqrt(2)*10.0
phi_vals_cos3 = a0 + R * np.cos(3*TH)
print(f"\n  phi(theta_i) = {phi_vals_cos3}")
print(f"  All equal: {np.allclose(phi_vals_cos3, phi_vals_cos3[0])}")

sq = np.sqrt(phi_vals_cos3)
Q_cos3 = np.sum(sq**2) / np.sum(sq)**2
print(f"\n  Q = sum(m_i) / (sum(sqrt(m_i)))^2 = {Q_cos3:.6f}")
print(f"  Expected Koide = 2/3 = {2/3:.6f}")
print(f"  Actual Q = 1/3 = {1/3:.6f}  (degenerate masses → Q=1/3)")

print()
print("=" * 60)
print("3. WHERE ARE THE MAXIMA OF cos(3θ)?")
print("=" * 60)
theta_fine = np.linspace(0, 2*np.pi, 10000)
phi_cos3 = np.cos(3 * theta_fine)

maxima_idx = np.where(np.diff(np.sign(np.diff(phi_cos3))) < 0)[0] + 1
minima_idx = np.where(np.diff(np.sign(np.diff(phi_cos3))) > 0)[0] + 1

print(f"\n  Maxima at θ = ", end="")
for idx in maxima_idx:
    print(f"{theta_fine[idx]:.4f} ({theta_fine[idx]*180/np.pi:.1f}°)", end="  ")

print(f"\n  Z₃ points:    ", end="")
for th in TH:
    print(f"{th:.4f} ({th*180/np.pi:.1f}°)", end="  ")

print(f"\n\n  Minima at θ = ", end="")
for idx in minima_idx:
    print(f"{theta_fine[idx]:.4f} ({theta_fine[idx]*180/np.pi:.1f}°)", end="  ")

print(f"""

  KEY RESULT:
  Maxima of cos(3θ) are EXACTLY at the Z₃ points (0°, 120°, 240°).
  → Fermions localize at MAXIMA of scalar profile (largest Yukawa)
  → cos(3θ) naturally places fermions at Z₃ points!
""")

print("=" * 60)
print("4. COMBINED PROFILE: a0 + R1*cos(θ) + R3*cos(3θ)")
print("=" * 60)
print("""
  At Z₃ points:
  phi(θ_i) = a0 + R1*cos(θ_i + φ') + R3*cos(3θ_i + φ₃)
           = a0 + R1*cos(θ_i + φ') + R3*cos(φ₃)   [since cos(3θ_i)=const]
           = [a0 + R3*cos(φ₃)] + R1*cos(θ_i + φ')
           =       A0           + R1*cos(θ_i + φ')

  This is IDENTICAL to the original model with a0 → A0!
  → cos(3θ) just shifts a0, does NOT change Koide structure.
  → Koide condition: R1 = √2 × A0 = √2 × (a0 + R3*cos(φ₃))
""")

# Numerical verification
a0, R1, R3 = 10.0, np.sqrt(2)*10.0, 3.0
phi3 = 0.5  # arbitrary phase for R3 term

phi_combined = a0 + R1*np.cos(TH) + R3*np.cos(3*TH + phi3)
phi_effective = (a0 + R3*np.cos(phi3)) + R1*np.cos(TH)

print("  Numeric check:")
print(f"  Combined profile at Z3:  {phi_combined}")
print(f"  Effective (shifted a0):  {phi_effective}")
print(f"  Identical: {np.allclose(phi_combined, phi_effective)}")

sq_comb = np.sqrt(np.abs(phi_combined))
Q_comb = np.sum(sq_comb**2) / np.sum(sq_comb)**2
A0 = a0 + R3*np.cos(phi3)
Q_target = 1/3 + R1**2/(6*A0**2)
print(f"\n  Q (combined) = {Q_comb:.6f}")
print(f"  Q (formula)  = {Q_target:.6f}")
print(f"  Q = 2/3 requires R1 = √2 × A0 = {np.sqrt(2)*A0:.4f} (current R1={R1:.4f})")

print()
print("=" * 60)
print("5. PHYSICAL PICTURE")
print("=" * 60)
print(f"""
  TWO ROLES of the scalar profile:

  cos(3θ) component  → GEOMETRIC ROLE
  ─────────────────────────────────────
  • 3 maxima exactly at Z₃ points
  • Fermions localize at maxima → Z₃ from geometry
  • All three generations get EQUAL base mass (no hierarchy)
  • Equivalent to shifting a₀

  cos(θ) component   → FLAVOR ROLE
  ─────────────────────────────────────
  • Different at each Z₃ point
  • Breaks the 3-fold degeneracy → mass hierarchy
  • Controls Koide relation (R₁ = √2 A₀)
  • Gives CKM/PMNS structure

  TOGETHER:
  ┌─────────────────────────────────────────────────────┐
  │  cos(3θ): WHY fermions are at Z₃ points            │
  │  cos(θ):  WHY they have different masses           │
  │  Koide:   still holds, with a₀ → A₀               │
  └─────────────────────────────────────────────────────┘
""")

print("=" * 60)
print("6. Z₃ QUANTIZATION: WHY k=3 AND NOT k=2 OR k=4?")
print("=" * 60)
print("""
  For cos(k×θ) evaluated at Z₃ points θ_i = 2π(i-1)/3:
  cos(k × 2π(i-1)/3)

  k mod 3 = 0  (k=3,6,...): CONSTANT → no flavor structure
                              → geometric localization role
  k mod 3 = 1  (k=1,4,...): different at each point
                              → flavor/hierarchy role
  k mod 3 = 2  (k=2,5,...): different at each point
                              → second harmonic (Koide correction)

  Only k ≡ 0 (mod 3) creates the symmetric 3-point structure.
  The SMALLEST such k is k=3 → cos(3θ) is the natural choice.
""")

# Verify for k=1,2,3
for k in [1, 2, 3, 4, 6]:
    vals = np.cos(k * TH)
    is_const = np.allclose(vals, vals[0])
    print(f"  k={k}: cos({k}θ) at Z₃ = {vals.round(4)} "
          f"→ {'CONSTANT (geometric)' if is_const else 'variable (flavor)'}")

print()
print("=" * 60)
print("7. GAUSSIAN WIDTH FROM CONE GEOMETRY")
print("=" * 60)
print("""
  Near a maximum of cos(3θ) at θ=0, Taylor expand:
  cos(3θ) ≈ 1 - (3θ)²/2 = 1 - 9θ²/2

  The scalar profile near θ=0:
  phi(θ) ≈ (a₀+R₃) - 9R₃θ²/2

  If fermion mass ∝ phi(θ), the Gaussian zero mode width:
  σ² ~ 1/(9R₃/2) = 2/(9R₃)

  → σ = √(2/9R₃) = √2 / (3√R₃)

  This connects the Gaussian width σ to the R₃ amplitude!
  σ is no longer free — it follows from the geometry.
""")

R3_vals = [1.0, 3.0, 5.0, 10.0, 50.0]
print(f"  {'R₃':>8}  {'σ_predicted':>14}  {'comment'}")
for r3 in R3_vals:
    sigma_pred = np.sqrt(2) / (3 * np.sqrt(r3))
    comment = "<-- near quark σ=0.944" if abs(sigma_pred - 0.944) < 0.1 else ""
    print(f"  {r3:>8.1f}  {sigma_pred:>14.4f}  {comment}")

# Find R3 that gives sigma = 0.944
sigma_target = 0.944
R3_for_sigma = 2 / (9 * sigma_target**2)
print(f"\n  For σ = {sigma_target} (quark fit value):")
print(f"  R₃ = 2/(9σ²) = {R3_for_sigma:.4f}")
print(f"  Verify: σ = √2/(3√R₃) = {np.sqrt(2)/(3*np.sqrt(R3_for_sigma)):.4f} ✓")

print()
print("=" * 60)
print("SUMMARY")
print("=" * 60)
print(f"""
  cos(3θ) profile analysis:

  ✓ Maxima exactly at Z₃ points → localizes fermions there
  ✓ Does NOT destroy Koide relation (just shifts a₀)
  ✓ Gaussian width σ follows from R₃ amplitude: σ = √2/(3√R₃)
  ✓ k=3 is the MINIMAL harmonic with 3-fold symmetry
  ✗ Alone gives Q=1/3 (degenerate masses) — needs cos(θ) too

  Combined profile phi = a₀ + R₁cos(θ) + R₃cos(3θ):
  • R₃ controls WHERE fermions localize (geometry → Z₃)
  • R₁ controls HOW MUCH they differ in mass (hierarchy)
  • Koide: R₁ = √2 (a₀ + R₃cos(φ₃))
  • σ ~ √2/(3√R₃)  [no longer free parameter!]

  R₃ for quark σ=0.944: R₃ ≈ {R3_for_sigma:.3f}
  This is a NEW CONSTRAINT linking σ and the scalar profile.
""")
```
